```python
import requests
import pandas as pd
```

```python
def get_universities(country_name):
    base_url = "http://universities.hipolabs.com/search"
    params = {'country': country_name}
    response = requests.get(base_url, params = params)
    data = response.json()
    df = pd.DataFrame(data)
    print(df)
```

```python
get_universities("Nepal")
```

```python
# live api

import requests
import pandas as pd

def fetch_bitget_price(symbol="BTCUSDT"):
    url = f"https://api.bitget.com/api/v2/spot/market/tickers?symbol={symbol}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        # the data is inside the 'data' key as a list
        ticker_list = data.get('data', [])
        
        df = pd.DataFrame(ticker_list)
        
        return df[['symbol', 'lastPr', 'high24h', 'low24h', 'ts']]
    
    else:
        return "Error fetching data"

print(fetch_bitget_price("BTCUSDT"))
```
