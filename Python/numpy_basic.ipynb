# Titanic Data — NumPy for Beginners

**What is NumPy?**
NumPy is a calculator for lists of numbers.
Pandas is great for tables. NumPy is great for doing math on numbers quickly.

**Real-world example:**
Imagine you have a list of all eSewa transaction amounts for the month.
NumPy can calculate the total, average, highest, lowest — instantly, even for millions of numbers.

# PART 1: NUMPY BASICS — Learn with Simple Examples

```python
# Step 1: Import NumPy
# np is the short nickname everyone uses for numpy

import numpy as np
import pandas as pd

print('NumPy imported successfully!')
print('NumPy version:', np.__version__)
```

```python
# Step 2: Create a simple NumPy array
# A regular Python list:  [10, 20, 30, 40, 50]
# A NumPy array is similar but much faster for calculations

# Example: Daily sales (in thousands Rs.) for a shop in Kathmandu
daily_sales = np.array([15, 22, 18, 30, 27, 19, 25])
 
print('Daily sales array:', daily_sales)
print('Type:', type(daily_sales))
```

```python
# Step 3: Basic math operations on arrays
# NumPy applies the math to ALL values at once — no need for loops

print('Original sales:', daily_sales)
print()
print('Sales + 5 (bonus added):',     daily_sales + 5)
print('Sales * 2 (doubled):',         daily_sales * 2)
print('Sales - 3 (discount applied):', daily_sales - 3)
print('Sales / 5 (divided):',         daily_sales / 5)
```

```python
# Step 4: Important NumPy calculations
# These are the most used functions in analytics

print('Sum of all sales:    ', np.sum(daily_sales))
print('Average sale:        ', np.mean(daily_sales))
print('Median sale:         ', np.median(daily_sales))
print('Highest sale:        ', np.max(daily_sales))
print('Lowest sale:         ', np.min(daily_sales))
print('Standard deviation:  ', round(np.std(daily_sales), 2))
# Standard deviation = how much values spread from the average
# Low std = consistent sales | High std = up and down sales
```

```python
# Step 5: Sorting and finding positions

scores = np.array([85, 92, 78, 95, 88, 76, 91])

print('Original scores:         ', scores)
print('Sorted (low to high):    ', np.sort(scores))
print('Sorted (high to low):    ', np.sort(scores)[::-1])
print('Position of highest:     ', np.argmax(scores))  # which index is highest
print('Position of lowest:      ', np.argmin(scores))  # which index is lowest
```

```python
# Step 6: Percentiles — dividing data into parts
# 25th percentile = 25% of values are below this
# 50th percentile = median (middle)
# 75th percentile = 75% of values are below this
# Used in salary analysis, test scores, business benchmarks

# 25th percentile
# → You are ahead of 25 people
# → 25% of people earn less than you
# 50th percentile (median)
# → You are in the middle
# → Half earn less, half earn more
# 75th percentile
# → You are ahead of 75 people
# → Most people earn less than you, only top 25% earn more

salaries = np.array([25000, 35000, 28000, 55000, 42000, 31000, 48000, 22000, 60000, 38000])

print('Salary data (Rs.):', salaries)
print()
print('25th percentile (bottom quarter):', np.percentile(salaries, 25))
print('50th percentile (middle/median): ', np.percentile(salaries, 50))
print('75th percentile (top quarter):   ', np.percentile(salaries, 75))
```

```python
# Step 7: Filtering with conditions
# We can ask: which values are greater than X?

ages = np.array([25, 34, 17, 45, 12, 60, 28, 15, 72, 38])

# Which ages are adults (18 and above)?
adult_filter = ages >= 18
print('Age filter (True = adult):', adult_filter)
print()

# Get only the adult ages
adult_ages = ages[adult_filter]
print('Adult ages only:', adult_ages)
```

```python
# Compare two array

ages = np.array([25, 34, 17, 45])
limits = np.array([20, 30, 18, 40])

result = ages >= limits
print(result)

# If you want 1st with 3rd, just index:

ages[0] >= limits[2]
```

```python
# Step 8: Count and percentage from conditions

ages = np.array([25, 34, 17, 45, 12, 60, 28, 15, 72, 38])

total = len(ages)
adults = np.sum(ages >= 18)         # counts True values (True = 1, False = 0)
children = np.sum(ages < 18)

print(f'Total people : {total}')
print(f'Adults       : {adults} ({(adults/total)*100:.0f}%)')
print(f'Children     : {children} ({(children/total)*100:.0f}%)')
```

```python
# Step 9: np.where() — like an IF condition in Excel
# Format: np.where(condition, value if True, value if False)

marks = np.array([45, 72, 38, 85, 60, 55, 90, 40, 78, 62])

# If marks >= 50, write 'Pass', else write 'Fail'
result = np.where(marks >= 50, 'Pass', 'Fail')

print('Marks:', marks)
print('Result:', result)
```

```python
# Step 10: np.unique() — find all unique values

cities = np.array(['Kathmandu', 'Pokhara', 'Kathmandu', 'Lalitpur', 'Pokhara', 'Kathmandu'])

print('All cities:', cities)
print('Unique cities:', np.unique(cities))

# Also count how many times each appears
unique_cities, counts = np.unique(cities, return_counts=True)
# for city, count in zip(unique_cities, counts):  # zip() combines multiple lists (or arrays) element-by-element
#     print(f'{city}: {count} times')

print(unique_cities,counts)
```

# PART 2: NUMPY WITH TITANIC DATA

Now we use NumPy on the real Titanic dataset.

```python
# Load and clean the Titanic data first
df = pd.read_csv('titanic_cleaned.csv')
# df['Age']  = df['Age'].fillna(df['Age'].median()).round(0)
# df['Fare'] = df['Fare'].fillna(df['Fare'].median()).round(2)

print('Data loaded. Shape:', df.shape)
```

```python
df['Age'].head()
```

```python
# Convert Pandas columns to NumPy arrays for fast calculation
# .values or .to_numpy() extracts the numbers from a pandas column

ages      = df['Age'].to_numpy() # Convert it into NumPy array
fares     = df['Fare'].to_numpy()
survived  = df['Survived'].to_numpy()

print('Age array (first 10):', ages[:10])
print('Fare array (first 10):', fares[:10])
print('Survived array (first 10):', survived[:10])
```

```python
# Age statistics using NumPy
print('=== AGE STATISTICS ===')
print(f'Youngest passenger : {np.min(ages):.0f} years')
print(f'Oldest passenger   : {np.max(ages):.0f} years')
print(f'Average age        : {np.mean(ages):.1f} years')
print(f'Median age         : {np.median(ages):.0f} years')
print(f'Std deviation      : {np.std(ages):.1f} years')
print()
print('Interpretation: Std of', round(np.std(ages),1), 'means ages vary by about that much from the average')
```

```python
# Fare statistics using NumPy
print('=== FARE (TICKET PRICE) STATISTICS ===')
print(f'Cheapest ticket : {np.min(fares):.2f}')
print(f'Most expensive  : {np.max(fares):.2f}')
print(f'Average fare    : {np.mean(fares):.2f}')
print(f'Median fare     : {np.median(fares):.2f}')
print(f'Total revenue   : {np.sum(fares):.2f}')
print()
# Note: Average is much higher than median — a few very expensive tickets pull the average up
print('Note: Mean is higher than median because a few luxury tickets are very expensive')
```

```python
# Survival statistics using NumPy
# Survived = 1 means survived. So average of 0s and 1s = survival rate

print('=== SURVIVAL STATISTICS ===')
total      = len(survived)
n_survived = np.sum(survived)           # sum of 1s = number who survived
n_died     = total - n_survived
rate       = np.mean(survived) * 100    # average of 0s and 1s = %

print(f'Total passengers : {total}')
print(f'Survived         : {n_survived}')
print(f'Died             : {n_died}')
print(f'Survival rate    : {rate:.1f}%')
```

```python
# Age percentiles — dividing passengers into age groups
print('=== AGE PERCENTILES ===')
p25 = np.percentile(ages, 25)
p50 = np.percentile(ages, 50)
p75 = np.percentile(ages, 75)

print(f'25th percentile : {p25:.0f} years  (25% of passengers were younger than this)')
print(f'50th percentile : {p50:.0f} years  (half were younger, half were older)')
print(f'75th percentile : {p75:.0f} years  (75% of passengers were younger than this)')
```

```python
# Filter using NumPy conditions — survival rate for different groups

# Get arrays for male and female passengers
sex_array      = df['Sex'].to_numpy()
survived_array = df['Survived'].to_numpy()

# Create True/False filters
female_filter = sex_array == 'female'
male_filter   = sex_array == 'male'

# Apply filters to survival array
female_survival_rate = np.mean(survived_array[female_filter]) * 100
male_survival_rate   = np.mean(survived_array[male_filter]) * 100

print(f'Female survival rate : {female_survival_rate:.1f}%')
print(f'Male survival rate   : {male_survival_rate:.1f}%')
```

```python
# Use np.where() to create a new column — like Excel IF
# If Survived=1 write 'Survived', else write 'Died'

df['Status'] = np.where(df['Survived'] == 1, 'Survived', 'Died')

print('Status column added:')
print(df['Status'].value_counts())
```

```python
# Use np.where() to create AgeGroup
# Note: np.where only handles 2 conditions (like if/else)
# For 3 conditions, we nest them

df['AgeGroup'] = np.where(
    df['Age'] < 18, 'Child',
    np.where(df['Age'] < 60, 'Adult', 'Senior')
)

print('Age group distribution:')
print(df['AgeGroup'].value_counts())
```

```python
# Count passengers by age group using NumPy
# group → the value to print (like "Adult")
    # : → start of formatting rules
    # 8 → minimum width = 8 characters

# {group:<8} = left alignment
# {group:>8} = right alignment
# {group:^8} = center alignment

age_groups      = df['AgeGroup'].to_numpy()
unique_groups, group_counts = np.unique(age_groups, return_counts=True)

print('Age group count using NumPy:')
for group, count in zip(unique_groups, group_counts):
    pct = count / len(age_groups) * 100
    print(f'  {group:8} : {count} passengers ({pct:.1f}%)')

# Zip works as:
#         | unique_groups | group_counts |
#         | ------------- | ------------ |
#         | Child         | 50           |
#         | Adult         | 300          |
#         | Senior        | 80           |


# So zip() turns them into pairs:
#         ('Child', 50)
#         ('Adult', 300)
#         ('Senior', 80)
```

```python
# Fare analysis by class using NumPy filtering
pclass_array = df['Pclass'].to_numpy()
fare_array   = df['Fare'].to_numpy()

for cls in [1, 2, 3]:
    filter_cls = pclass_array == cls         # True/False filter,  Let’s say cls = 1 so filter_cls will keep True where cls == pclass_array
    class_fares = fare_array[filter_cls]     # apply filter
    print(f'Class {cls} — Avg Fare: {np.mean(class_fares):.2f} | Max: {np.max(class_fares):.2f} | Min: {np.min(class_fares):.2f}')
```

```python
# Calculate cumulative sum of fares
# Imagine you are collecting ticket payments one by one — cumsum shows running total

sorted_fares  = np.sort(fares)                 # sort from lowest to highest
running_total = np.cumsum(sorted_fares)        # running total like [10, 20, 50, 100] -> [[10, 30, 80, 180]

print('First 5 fares (sorted):', sorted_fares[:5].round(2))
print('Running total after 5  :', running_total[:5].round(2))
print()
print(f'Total ticket revenue collected: {running_total[-1]:.2f}')
```

```python
# Clip values — set a maximum and minimum limit
# Like capping a salary at Rs. 50 per unit for reporting purposes
# np.clip(array, min_value, max_value)

fares_capped = np.clip(fares, 0, 100)     # cap all fares at 100 (remove extreme values)

# Anything above 100 becomes 100
# Anything below 0 becomes 0
# Everything else stays same

print(f'Original max fare  : {np.max(fares):.2f}')
print(f'Capped max fare    : {np.max(fares_capped):.2f}')
print(f'Original avg fare  : {np.mean(fares):.2f}')
print(f'Capped avg fare    : {np.mean(fares_capped):.2f}')
# Capping removes outlier effect on the average
```

```python
# Normalize fares between 0 and 1
# Normalization is used before machine learning — makes all numbers on same scale
# Formula: (value - min) / (max - min)

fares_normalized = (fares - np.min(fares)) / (np.max(fares) - np.min(fares))

print('First 5 original fares:', fares[:5].round(2))
print('First 5 normalized (0-1):', fares_normalized[:5].round(4))
print()
print(f'Normalized min: {np.min(fares_normalized):.2f}')   # should be 0
print(f'Normalized max: {np.max(fares_normalized):.2f}')   # should be 1
```

```python
# Final NumPy summary

print('='*45)
print('   NUMPY ANALYSIS SUMMARY')
print('='*45)
print(f'Total Passengers       : {len(ages)}')
print(f'Age    — Mean/Median   : {np.mean(ages):.1f} / {np.median(ages):.0f}')
print(f'Fare   — Mean/Median   : {np.mean(fares):.2f} / {np.median(fares):.2f}')
print(f'Total Revenue          : {np.sum(fares):.2f}')
print(f'Survival Rate          : {np.mean(survived)*100:.1f}%')
print(f'Age Std Deviation      : {np.std(ages):.1f} years')
print(f'Fare Std Deviation     : {np.std(fares):.2f}')
print('='*45)
```

| Function | What it does | Real-world example |

| `np.max()` | Highest value | Best exam score |
| `np.min()` | Lowest value | Minimum temperature |
| `np.std()` | How spread out values are | Income inequality |
| `np.sort()` | Sort values | Rank students |
| `np.percentile()` | Divide into % groups | Top 25% earners |
| `np.where()` | IF condition | Pass/Fail labeling |
| `np.unique()` | Find all unique values | List of cities |
| `np.cumsum()` | Running total | Monthly balance |
| `np.clip()` | Set min/max limit | Cap extreme values |
