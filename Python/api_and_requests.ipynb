# 🌐 API & Requests Library in Python

**What is an API?**

API stands for **Application Programming Interface**.

Think of it like a **waiter in a restaurant**:
- You (the customer) = your Python code
- The kitchen = a server with data
- The waiter = the API

You tell the waiter what you want → the waiter goes to the kitchen → brings back your food (data).

**Real examples of APIs:**
- Weather app on your phone gets weather data from an API
- eSewa checks your bank balance through an API
- Daraz shows product prices through an API
- Google Maps shows directions through an API

## 1. 📦 What is the `requests` Library?

The `requests` library is a Python tool that lets you **talk to APIs** — send requests and receive responses — in just a few lines of code.

**Types of requests (HTTP Methods):**

| Method | What it does | Real-life example |
|--------|-------------|-------------------|
| `GET` | **Read / Get** data | Open a webpage, get weather info |
| `POST` | **Send / Create** data | Submit a form, place an order |
| `PUT` | **Update** existing data | Edit your profile |
| `DELETE` | **Delete** data | Delete an account |

> For beginners, we mostly use **GET** (getting data) and **POST** (sending data).

## 3. 🚀 Making Your First API Request (GET)

We will use a **free public API** — no account needed!

**JSONPlaceholder** is a free fake API for practice. It's like a dummy database.

URL: `https://jsonplaceholder.typicode.com`

````python
# Make a GET request — like opening a URL in your browser

url = "https://jsonplaceholder.typicode.com/posts/1"

response = requests.get(url)

print(response)
print(f"Request sent to : {url}")
print(f"Status Code     : {response.status_code}")
````

---
## 4. 📬 Understanding the Response

When you send a request, you get back a **response**. It contains:

| Part | What it means |
|------|---------------|
| `status_code` | Was the request successful? |
| `text` | Raw response as plain text |
| `json()` | Response as Python dictionary |
| `headers` | Technical info about the response |

**Common Status Codes:**

| Code | Meaning |
|------|---------|
| `200` | ✅ Success — everything is fine |
| `201` | ✅ Created — data was created |
| `400` | ❌ Bad Request — your request has an error |
| `401` | ❌ Unauthorized — you need to log in |
| `404` | ❌ Not Found — the URL doesn't exist |
| `500` | ❌ Server Error — the server has a problem |

````python
url = "https://jsonplaceholder.typicode.com/posts/1"
response = requests.get(url)

# Status code
print(f"Status Code : {response.status_code}")

# Check if successful
if response.status_code == 200:
    print("✅ Request was successful!")
else:
    print("❌ Something went wrong!")

# Raw text response
print(f"\nRaw Response:\n{response.text}")
````

---
## 5. 📄 Working with JSON Data

**JSON** (JavaScript Object Notation) is the most common format APIs use to send data.

It looks exactly like a Python dictionary!

```json
{"name": "Ram Thapa", "city": "Kathmandu", "salary": 45000}
```

Use `response.json()` to convert the response into a Python dictionary you can work with.

````python
url = "https://jsonplaceholder.typicode.com/posts/1"
response = requests.get(url)

# Convert response to Python dictionary
data = response.json()

print("Type:", type(data))    # It's a dict!
print()
print("Full data:")
print(data)
````

````python
# Access individual fields just like a dictionary

print(f"Post ID   : {data['id']}")
print(f"User ID   : {data['userId']}")
print(f"Title     : {data['title']}")
print(f"Body      : {data['body'][:60]}...")   # first 60 characters
````

````python
snippet = data['body'][:60]

print(snippet)
print("Length:", len(snippet))
print("Raw view:", repr(snippet))
````

## 6. 📤 POST Request — Sending Data to an API

**GET** = asking for data (reading)

**POST** = sending data (creating something new)

Real examples:
- Submitting a form on a website
- Placing an order on Daraz
- Registering a new user

````python
# POST Request — Create a new post

url = "https://jsonplaceholder.typicode.com/posts"

# Data we want to SEND
new_post = {
    "title" : "API Course",
    "body"  : "This course teaches Python's API.",
    "userId": 1
}

# Send POST request with our data
response = requests.post(url, json=new_post)  # keep the data in JSON format

print(f"Status Code : {response.status_code}")  # 201 = Created successfully
print(f"Response    :")
print(response.json())
````

| Format    | requests parameter | What you pass |
| --------- | ------------------ | ------------- |

| XML       | `data=`            | string        |
| File      | `files=`           | file object()  |

files= when you are uploading an actual file from disk

data= when you are sending raw content, not a file.
