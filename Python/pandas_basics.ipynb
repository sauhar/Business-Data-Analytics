## 2. 📊 Series — A Single Column

A **Series** is like a single column in Excel — it holds one list of values with an index (row number).

```
Index    Value
  0      12000
  1      15000
  2       9800
```

````python
# Create a simple Series from a list

daily_sales = pd.Series([12000, 15000, 9800, 22000, 18300])
print(daily_sales)
````

````python
# Series with custom index labels

branch_sales = pd.Series(
    [150000, 210000, 98000, 175000],
    index=["New Road", "Thamel", "Patan", "Bhaktapur"]
)

print(branch_sales)
print()

# Access by label
print(f"Thamel sales: Rs. {branch_sales['Thamel']:,}")

# Basic stats on a Series
print(f"\nTotal  : Rs. {branch_sales.sum():,}")
print(f"Average: Rs. {branch_sales.mean():,.0f}")
print(f"Highest: Rs. {branch_sales.max():,}")
````

## 3. 🗃️ DataFrame — A Full Table

A **DataFrame** is a full table with **rows and columns** — just like an Excel spreadsheet.

```
   Name         City       Salary
0  Ram Thapa    Kathmandu  45000
1  Sita Rai     Pokhara    52000
2  Hari Magar   Lalitpur   61000
```

- Each **column** is a Series
- All columns together make a DataFrame

## 4. ➕ Creating a DataFrame

````python
# Method 1: From a dictionary (most common way)
# Each key = column name, each value = list of data

data = {
    "Name"      : ["Ram Thapa", "Sita Rai", "Hari Magar", "Gita Shrestha", "Bikash KC"],
    "City"      : ["Kathmandu", "Pokhara", "Lalitpur", "Biratnagar", "Kathmandu"],
    "Department": ["Sales", "Finance", "IT", "HR", "Sales"],
    "Salary"    : [45000, 52000, 78000, 41000, 49000],
    "Experience": [3, 5, 7, 2, 4]
}

df = pd.DataFrame(data)
print(df)
````

````python
# Method 2: From a list of dictionaries
# Each dictionary = one row

products = [
    {"Product": "Laptop",    "Brand": "HP",      "Price": 85000, "Stock": 10},
    {"Product": "Mobile",    "Brand": "Samsung",  "Price": 45000, "Stock": 25},
    {"Product": "Tablet",    "Brand": "Xiaomi",   "Price": 35000, "Stock": 8},
    {"Product": "Headphone", "Brand": "JBL",      "Price": 8000,  "Stock": 40},
]

df_products = pd.DataFrame(products)
print(df_products)
````

## 5. 👀 Viewing Data

When you have a large dataset, you don't want to see all rows at once. These commands help you quickly explore the data.

## 7. 🔍 Filtering Data

Filtering = showing only rows that match a condition.
This is like the **Filter** button in Excel!

````python
# Filter: Only employees in Kathmandu
# First df["City"] → select the column to compare
# Second df[...] → select the rows from the DataFrame that satisfy the condition

df[df["City"] == "Kathmandu"]
df["City"] == "Kathmandu"
````

````python
# Filter: Only employees with salary above 50,000
df[df["Salary"] > 50000]
````

````python
# Filter with TWO conditions
# & means AND  |  means OR

# Employees in Sales department AND salary > 45000
df[(df["Department"] == "Sales") & (df["Salary"] > 45000)]
````

````python
# Filter: Employees in Kathmandu OR Pokhara
df[(df["City"] == "Kathmandu") | (df["City"] == "Pokhara")]
````

````python
# isin() — filter by multiple values (cleaner way)
df[df["City"].isin(["Kathmandu", "Pokhara", "Lalitpur"])]
````

## 8. ➕ Adding & Removing Columns

````python
# Add a new column — Annual Salary
df["Annual Salary"] = df["Salary"] * 12
df[["Name", "Salary", "Annual Salary"]].head()
````

````python
# Add a column based on a condition
# If experience >= 5 years → Senior, else → Junior

df["Level"] = df["Experience"].apply(lambda x: "Senior" if x >= 5 else "Junior")
df[["Name", "Experience", "Level"]].head(8)
````

````python
# Add bonus column — 10% of salary for Senior, 5% for Junior
df["Bonus"] = df.apply(
    lambda row: row["Salary"] * 0.10 if row["Level"] == "Senior" else row["Salary"] * 0.05,
    axis=1
)
df[["Name", "Salary", "Level", "Bonus"]].head(8)
````

````python
# Remove a column using drop()
df_clean = df.drop(columns=["Annual Salary", "Bonus"])
df_clean.head()
````

## 11. 🚫 Handling Missing Data

Real-world data always has missing values. Pandas makes it easy to find and fix them.

````python
import numpy as np

# Create a dataset with missing values (NaN = Not a Number = empty cell)
data_missing = {
    "Name"  : ["Ram", "Sita", "Hari", "Gita", "Bikash"],
    "City"  : ["Kathmandu", None, "Lalitpur", "Biratnagar", None],
    "Salary": [45000, 52000, np.nan, 41000, 49000],
    "Phone" : ["98XXXXXXXX", "97XXXXXXXX", None, None, "98XXXXXXXX"]
}

df_missing = pd.DataFrame(data_missing)
df_missing
````

````python
# isnull() — show which cells are missing
df_missing.isnull()
````

````python
# Count missing values per column
print(df_missing.isnull().sum())
````

````python
# Option 1: Fill missing values with a default value
df_filled = df_missing.copy()
df_filled["City"]   = df_filled["City"].fillna("Unknown")
df_filled["Salary"] = df_filled["Salary"].fillna(df_filled["Salary"].mean())  # fill with average
df_filled["Phone"]  = df_filled["Phone"].fillna("Not Provided")
df_filled
````

````python
# Option 2: Drop rows that have ANY missing value
df_dropped = df_missing.dropna()
df_dropped
````

## 12. 📦 Grouping & Aggregation

**groupby()** is like a Pivot Table in Excel — it groups rows by a category and calculates totals, averages, etc.

````python
# Average salary by Department
df.groupby("Department")["Salary"].mean().round(0)
````

````python
# Total salary cost per Department
df.groupby("Department")["Salary"].sum()
````

````python
# Multiple statistics at once using agg()
df.groupby("Department")["Salary"].agg(["count", "mean", "min", "max"]).round(0)
````

````python
# Group by City — count employees and average salary
df.groupby("City").agg(
    Employee_Count=("Name", "count"),
    Avg_Salary=("Salary", "mean"),
    Total_Salary=("Salary", "sum")
).round(0)
````

## 13. 📂 Reading a CSV File

In real business, data comes from CSV files (Excel exports, database exports, etc.).
Pandas can read them in one line!

````python
# First, let's CREATE a sample CSV file to practice with

sales_data = {
    "Date"      : ["2025-01-01", "2025-01-02", "2025-01-03", "2025-01-04", "2025-01-05",
                   "2025-01-06", "2025-01-07", "2025-01-08", "2025-01-09", "2025-01-10"],
    "Branch"    : ["Thamel", "Patan", "Thamel", "Bhaktapur", "Patan",
                   "Thamel", "Bhaktapur", "Patan", "Thamel", "Bhaktapur"],
    "Product"   : ["Laptop", "Mobile", "Tablet", "Laptop", "Headphone",
                   "Mobile", "Tablet", "Laptop", "Headphone", "Mobile"],
    "Quantity"  : [2, 5, 3, 1, 8, 4, 2, 3, 6, 7],
    "Unit_Price": [85000, 45000, 35000, 85000, 8000, 45000, 35000, 85000, 8000, 45000]
}

df_sales = pd.DataFrame(sales_data)
df_sales["Total"] = df_sales["Quantity"] * df_sales["Unit_Price"]

# Save to CSV
df_sales.to_csv("sales.csv", index=False)
print("✅ sales.csv file created!")
````

````python
# Now READ the CSV file back
df_from_csv = pd.read_csv("sales.csv")

print(f"Shape: {df_from_csv.shape}")
df_from_csv.head()
````

````python
# Analyze the CSV data

print("--- Sales Summary ---")
print(f"Total Revenue : Rs. {df_from_csv['Total'].sum():,}")
print(f"Total Items   : {df_from_csv['Quantity'].sum()}")
print(f"Avg Order     : Rs. {df_from_csv['Total'].mean():,.0f}")

print("\n--- Revenue by Branch ---")
print(df_from_csv.groupby("Branch")["Total"].sum().sort_values(ascending=False))
````

````python
# Save results to a new CSV
branch_report = df_from_csv.groupby("Branch")["Total"].sum().reset_index()
branch_report.columns = ["Branch", "Total_Revenue"]
branch_report.to_csv("branch_report.csv", index=False)
print("✅ branch_report.csv saved!")
print(branch_report)
````

## 14. 💼 Mini Business Project

**Scenario:** You are the Data Analyst at a Daraz Nepal office.
You have one month of order data. Your manager wants a complete sales report.

Tasks:
1. Load and explore the data
2. Find total revenue
3. Find best-selling product
4. Find top city by orders
5. Flag high-value orders (above Rs. 50,000)

````python
# Step 1: Create the dataset

orders = {
    "Order_ID" : [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Customer" : ["Ram", "Sita", "Hari", "Gita", "Bikash", "Priya", "Rohan", "Anita", "Suresh", "Mina"],
    "City"     : ["Kathmandu", "Pokhara", "Kathmandu", "Biratnagar", "Lalitpur",
                  "Kathmandu", "Pokhara", "Lalitpur", "Kathmandu", "Biratnagar"],
    "Product"  : ["Laptop", "Mobile", "Mobile", "Laptop", "Tablet",
                  "Headphone", "Laptop", "Mobile", "Tablet", "Headphone"],
    "Quantity" : [1, 2, 1, 1, 3, 5, 2, 1, 2, 4],
    "Price"    : [85000, 45000, 45000, 85000, 35000, 8000, 85000, 45000, 35000, 8000]
}

df_orders = pd.DataFrame(orders)
df_orders["Revenue"] = df_orders["Quantity"] * df_orders["Price"]

print("✅ Dataset created!")
df_orders
````

````python
# Step 2: Total Revenue
total_revenue = df_orders["Revenue"].sum()
total_orders = len(df_orders)
print(f"Total Orders  : {total_orders}")
print(f"Total Revenue : Rs. {total_revenue:,}")
````

````python
# Step 3: Best-selling product by revenue
product_revenue = df_orders.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
print("Revenue by Product:")
print(product_revenue)
````

````python
# Step 4: Top city by number of orders
city_orders = df_orders["City"].value_counts()
print("Orders by City:")
print(city_orders)
print(f"\n🏆 Top City: {city_orders.index[0]} with {city_orders.iloc[0]} orders")
````

````python
# Step 5: Flag high-value orders
df_orders["Order_Type"] = df_orders["Revenue"].apply(
    lambda x: "🔥 High Value" if x >= 50000 else "Regular"
)

print("All Orders with Type:")
df_orders[["Order_ID", "Customer", "Product", "Revenue", "Order_Type"]]
````

````python
# Final Report Summary
print("=" * 45)
print("   DARAZ NEPAL — MONTHLY SALES REPORT")
print("=" * 45)
print(f"Total Orders        : {total_orders}")
print(f"Total Revenue       : Rs. {total_revenue:,}")
print(f"Average Order Value : Rs. {df_orders['Revenue'].mean():,.0f}")
print(f"Top Product         : {product_revenue.index[0]}")
print(f"Top City            : {city_orders.index[0]}")
print(f"High Value Orders   : {(df_orders['Order_Type'] == '🔥 High Value').sum()}")
print("=" * 45)
````

## 🎉 Summary — Pandas Cheat Sheet

| Task | Code | Excel Equivalent |
|------|------|------------------|
| Create table | `pd.DataFrame(data)` | Type data in sheet |
| First 5 rows | `df.head()` | Scroll to top |
| Last 5 rows | `df.tail()` | Scroll to bottom |
| Shape | `df.shape` | Check row/col count |
| Column names | `df.columns` | Look at header row |
| Select column | `df["Name"]` | Click column |
| Select row | `df.iloc[0]` | Click row |
| Filter data | `df[df["City"] == "KTM"]` | Filter button |
| Add column | `df["New"] = ...` | Add new column |
| Delete column | `df.drop(columns=[...])` | Delete column |
| Sort | `df.sort_values("Col")` | Sort A→Z |
| Sum | `df["Col"].sum()` | SUM formula |
| Average | `df["Col"].mean()` | AVERAGE formula |
| Count unique | `df["Col"].value_counts()` | COUNTIF |
| Group by | `df.groupby("Col")` | Pivot Table |
| Read CSV | `pd.read_csv("file.csv")` | Open file |
| Save CSV | `df.to_csv("file.csv")` | Save As CSV |
| Missing values | `df.isnull().sum()` | Find blank cells |
| Fill missing | `df.fillna(value)` | Replace blanks |
