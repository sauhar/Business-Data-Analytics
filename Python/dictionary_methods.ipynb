# 📖 Dictionary Methods — All the Different Ways

---

A **dictionary** stores data in `key: value` pairs.

```python
customer = {"name": "Ram", "city": "Kathmandu", "age": 28}
```

In this notebook we will learn **all the different ways** to:
- **GET** — read values from a dictionary
- **PUT / UPDATE** — add or change values
- **REMOVE** — delete values
- **CHECK** — check if a key exists
- **LOOP** — go through a dictionary

---

---
# PART A: 📥 GET — Reading Values

There are **3 ways** to get a value from a dictionary.

| Method | Syntax | Safe? | Default Value? |
|--------|--------|-------|----------------|
| Square bracket | `dict["key"]` | ❌ Crashes if key missing | ❌ No |
| `.get(key)` | `dict.get("key")` | ✅ Safe | Returns `None` |
| `.get(key, default)` | `dict.get("key", "N/A")` | ✅ Safe | ✅ Your choice |

## A1. Way 1 — Square Bracket `dict["key"]`

The most basic way. Works fine **if the key exists**. But **crashes** if the key is missing.

````python
customer = {
    "name" : "Ram Thapa",
    "city" : "Kathmandu",
    "email": "ram@gmail.com"
}

# ✅ Works fine — key exists
print(customer["name"])
print(customer["city"])
````

````python
# ❌ CRASHES — key does not exist
try:
    print(customer["phone"])   # 'phone' is not in the dictionary
except KeyError as e:
    print(f"KeyError: {e} — this key does not exist!")
````

## A2. Way 2 — `.get(key)` — Safe, Returns None if Missing

````python
customer = {
    "name" : "Ram Thapa",
    "city" : "Kathmandu",
    "email": "ram@gmail.com"
}

# ✅ Key exists — returns the value
print(customer.get("name"))    # Ram Thapa

# ✅ Key missing — returns None (does NOT crash!)
print(customer.get("phone"))   # None
print(customer.get("age"))     # None
````

## A3. Way 3 — `.get(key, default_value)` — Safe, Returns YOUR Default

````python
customer = {
    "name" : "Ram Thapa",
    "city" : "Kathmandu",
}

# ✅ Key exists — returns the real value (ignores default)
print(customer.get("name",  "Unknown"))          # Ram Thapa
print(customer.get("city",   "Not found"))  

# ✅ Key missing — returns YOUR default value
print(customer.get("phone", "Not provided"))     # Not provided
print(customer.get("age",   0))                  # 0
print(customer.get("email", "No email found"))   # No email found
````

````python
# Real business use — eSewa transaction details

transaction = {
    "transaction_id" : "TXN-20240410-001",
    "sender"         : "Ram Thapa",
    "amount"         : 5000,
    "status"         : "Success"
    # 'receiver' and 'note' are missing!
}

print("--- eSewa Transaction Summary ---")
print(f"ID       : {transaction.get('transaction_id', 'N/A')}")
print(f"Sender   : {transaction.get('sender',         'Unknown')}")
print(f"Receiver : {transaction.get('receiver',       'Not specified')}")
print(f"Amount   : Rs. {transaction.get('amount',     0):,}")
print(f"Status   : {transaction.get('status',         'Pending')}")
print(f"Note     : {transaction.get('note',           'No note added')}")
````

# PART B: 📤 PUT / UPDATE — Adding & Changing Values

There are **4 ways** to add or update values.

| Method | Syntax | What it does |
|--------|--------|--------------|
| Square bracket assign | `dict["key"] = value` | Add new OR update existing |
| `.update({})` | `dict.update({"key": value})` | Update multiple keys at once |
| `.setdefault(key, value)` | `dict.setdefault("key", value)` | Add ONLY if key does NOT exist |
| Merge `\|=` (Python 3.9+) | `dict1 \|= dict2` | Merge two dictionaries |

## B1. Way 1 — Square Bracket `dict["key"] = value`

- If key **exists** → **updates** the value
- If key **doesn't exist** → **adds** it as new

````python
employee = {
    "name"      : "Sita Rai",
    "department": "Finance",
    "salary"    : 52000
}
print(f"Before: {employee}")

# Update existing key — salary changes
employee["salary"] = 58000
print(f"After salary update: {employee}")

# Add new key — 'city' didn't exist before
employee["city"] = "Pokhara"
print(f"After adding city: {employee}")
````

## B2. Way 2 — `.update({})` — Update Multiple Keys at Once

````python
employee = {
    "name"      : "Sita Rai",
    "department": "Finance",
    "salary"    : 52000
}
print(f"Before: {employee}")

# Update multiple keys in one line
employee.update({
    "salary"    : 60000,        # existing key — gets updated
    "city"      : "Lalitpur",   # new key — gets added
    "experience": 5,             # new key — gets added
    "age" : 32
})

print(f"After update: {employee}")
````

````python
# Real use — updating a product record in Daraz

product = {
    "name" : "Samsung Galaxy S24",
    "price": 85000,
    "stock": 20
}

# After a sale — update price and stock together
product.update({
    "price": 82000,   # price dropped
    "stock": 18,      # 2 items sold
    "discount": "5%"  # new field added
})

print(product)
````

## B3. Way 3 — `.setdefault(key, value)` — Add ONLY if Key is Missing

- If key **exists** → does **nothing**, returns existing value
- If key **doesn't exist** → **adds it** with the default value

Think of it as: *"Set this value, but only if there is nothing there already."*

````python
customer = {
    "name" : "Gita Shrestha",
    "city" : "Biratnagar"
}
print(f"Before: {customer}")

# Key 'city' EXISTS — setdefault does NOTHING (does not overwrite!)
customer.setdefault("city", "Unknown")
print(f"After setdefault city: {customer}")

# Key 'phone' does NOT exist — setdefault ADDS it
customer.setdefault("phone", "Not provided")
print(f"After setdefault phone: {customer}")

# Key 'membership' does NOT exist — setdefault ADDS it
customer.setdefault("membership", "Silver")
print(f"After setdefault membership: {customer}")
````

````python
# Real use — assign default membership level for new customers

customers = [
    {"name": "Ram",   "membership": "Gold"},
    {"name": "Sita"},                         # no membership key!
    {"name": "Hari",  "membership": "Silver"},
    {"name": "Gita"},                          # no membership key!
]

for c in customers:
    c.setdefault("membership", "Basic")   # add 'Basic' only if missing

for c in customers:
    print(f"{c['name']} → {c['membership']}")
````

## B4. Way 4 — Merge Two Dictionaries `|=` (Python 3.9+)

````python
# Basic employee info
basic_info = {
    "name" : "Bikash KC",
    "city" : "Kathmandu"
}

# Job info from another system
job_info = {
    "department": "Sales",
    "salary"    : 49000,
    "experience": 4
}

# Alternative — create a brand new merged dictionary
full_profile = basic_info | job_info
print(full_profile)
````

# PART C: 🗑️ REMOVE — Deleting Values

| Method | Syntax | What it does |
|--------|--------|--------------|
| `del` | `del dict["key"]` | Delete a key (crashes if missing) |
| `.pop(key)` | `dict.pop("key")` | Delete & return the value (crashes if missing) |
| `.pop(key, default)` | `dict.pop("key", "N/A")` | Delete & return value, or default if missing |
| `.popitem()` | `dict.popitem()` | Remove & return the LAST key-value pair |
| `.clear()` | `dict.clear()` | Remove ALL keys |

## C1. `del dict["key"]` — Simple Delete

````python
employee = {"name": "Priya", "city": "Pokhara", "salary": 61000, "temp_id": "T-009"}
print(f"Before: {employee}")

# Delete temp_id — no longer needed
del employee["temp_id"]
print(f"After del: {employee}")
````

````python
# ❌ Crashes if key doesn't exist
try:
    del employee["phone"]   # phone doesn't exist!
except KeyError as e:
    print(f"KeyError: {e} — cannot delete a key that doesn't exist")
````

## C2. `.pop(key)` — Delete AND Get the Value Back

````python
employee = {"name": "Rohan", "city": "Pokhara", "salary": 72000, "old_salary": 65000}
print(f"Before: {employee}")

# pop() removes AND returns the value
removed_value = employee.pop("city")

print(f"Removed value : {removed_value}")
print(f"After pop()   : {employee}")
````

## C3. `.pop(key, default)` — Safe Delete with Fallback

````python
employee = {"name": "Anita", "city": "Lalitpur", "salary": 38000}

# Key EXISTS — removes and returns value
city = employee.pop("city", "Not found")
print(f"Removed city  : {city}")
print(f"After pop     : {employee}")

# Key MISSING — does NOT crash, returns default
phone = employee.pop("phone", "No phone on record")
print(f"\nPhone result  : {phone}")
print(f"Dict unchanged: {employee}")
````

## C4. `.popitem()` — Remove Last Item

````python
product = {"name": "Laptop", "brand": "HP", "price": 85000, "stock": 10}
print(f"Before: {product}")

# Removes and returns the LAST key-value pair as a tuple
last_item = product.popitem()
print(f"Removed: {last_item}")
print(f"After  : {product}")
````

## C5. `.clear()` — Remove Everything

````python
cart = {"item1": "Laptop", "item2": "Charger", "item3": "Bag"}
print(f"Before clear: {cart}")

cart.clear()   # empty the cart after checkout
print(f"After clear : {cart}")
````

# PART D: 🔍 CHECK — Does a Key Exist?

| Method | Syntax | Returns |
|--------|--------|---------|
| `in` | `"key" in dict` | `True` or `False` |
| `not in` | `"key" not in dict` | `True` or `False` |
| `.keys()` | `dict.keys()` | All key names |
| `.values()` | `dict.values()` | All values |
| `.items()` | `dict.items()` | All key-value pairs |

````python
customer = {"name": "Ram", "city": "Kathmandu", "email": "ram@gmail.com"}

# Check if key exists
print("name" in customer)     # True
print("phone" in customer)    # False
print("phone" not in customer) # True
````

````python
# Use 'in' before accessing — safe pattern

if "email" in customer:
    print(f"Email: {customer['email']}")
else:
    print("No email on record")

if "phone" in customer:
    print(f"Phone: {customer['phone']}")
else:
    print("No phone on record")
````

````python
employee = {"name": "Sita", "department": "Finance", "salary": 52000}

# .keys() — all key names
print("Keys   :", list(employee.keys()))

# .values() — all values
print("Values :", list(employee.values()))

# .items() — all key-value pairs
print("Items  :", list(employee.items()))
````

# PART E: 🔁 LOOP — Going Through a Dictionary

| Way | What you get |
|-----|--------------|
| `for key in dict` | Only keys |
| `for key in dict.keys()` | Only keys (same as above) |
| `for val in dict.values()` | Only values |
| `for key, val in dict.items()` | Both keys AND values ✅ (most useful) |

````python
product = {"name": "Laptop", "brand": "HP", "price": 85000, "stock": 10}

# Loop through keys only
print("--- Keys ---")
for key in product:
    print(key)
````

````python
# Loop through values only
print("--- Values ---")
for value in product.values():
    print(value)
````

````python
# Loop through BOTH keys and values — most useful!
print("--- Key : Value ---")
for key, value in product.items():
    print(f"{key:<10} : {value}")
````

````python
# Real business use — print a customer profile card

customer = {
    "Name"       : "Priya Shrestha",
    "City"       : "Lalitpur",
    "Phone"      : "9841-XXXXXX",
    "Membership" : "Gold",
    "Total Spent": "Rs. 1,25,000"
}

print("=" * 35)
print("   CUSTOMER PROFILE — eSewa")
print("=" * 35)
for key, value in customer.items():
    print(f"  {key:<14}: {value}")
print("=" * 35)
````

# PART F: 🔄 COPY — Making a Copy of a Dictionary

⚠️ **Important!** If you do `dict2 = dict1`, both point to the **same** dictionary.
Changing one changes the other!

Always use `.copy()` to make a **separate** copy.

````python
# ❌ Wrong way — both variables point to SAME dictionary
original = {"name": "Ram", "salary": 45000}
wrong_copy = original          # NOT a real copy!

wrong_copy["salary"] = 99999   # changes BOTH!
print(f"Original   : {original}")    # original also changed!
print(f"Wrong copy : {wrong_copy}")
````

````python
# ✅ Right way — use .copy()
original = {"name": "Ram", "salary": 45000}
real_copy = original.copy()     # REAL separate copy

real_copy["salary"] = 99999     # only changes real_copy
print(f"Original  : {original}")    # unchanged ✅
print(f"Real copy : {real_copy}")
````

# PART G: 🏆 Complete Summary — Side by Side Comparison

````python
# Full demo — all methods in one place

employee = {
    "name"      : "Bikash Gurung",
    "department": "IT",
    "salary"    : 82000,
    "city"      : "Kathmandu"
}

print("ORIGINAL:")
print(employee)

print("\n--- GET ---")
print(employee["name"])                      # bracket — direct
print(employee.get("phone"))                 # get — returns None
print(employee.get("phone", "Not listed"))   # get with default

print("\n--- PUT / UPDATE ---")
employee["salary"] = 88000                   # update existing
employee["experience"] = 6                   # add new
employee.update({"city": "Lalitpur", "level": "Senior"})  # update multiple
employee.setdefault("bonus", 5000)           # add only if missing
employee.setdefault("city", "Unknown")       # does nothing (city exists)
print(employee)

print("\n--- REMOVE ---")
removed = employee.pop("level", "N/A")       # safe pop
print(f"Removed 'level': {removed}")
del employee["experience"]                   # direct delete
print(employee)

print("\n--- CHECK ---")
print("salary" in employee)                  # True
print("phone" in employee)                   # False
print(list(employee.keys()))
print(list(employee.values()))

print("\n--- LOOP ---")
for key, value in employee.items():
    print(f"  {key}: {value}")
````

```python
d = {"name": "Ram", "city": "Kathmandu"}

for key in d:                        # keys only

for key, val in d.items():           # both ✅ most useful
