```python
age_groups = df['AgeGroup'].to_numpy()

unique_groups, group_counts = np.unique(age_groups, return_counts=True)

print('Age group count using Numpy:')

for group, count in zip(unique_groups, group_counts):
    passengers_count = count / len(age_groups) * 100
    print(f'{group:8} : {count} passengers ({passengers_count:.1f}%)')
```

Fare analysis as per pclass using numpy filter


class 1 - agv fare: 65.25 | max: 512.25 | min:0.00

```python
pclass_array = df['Pclass'].to_numpy()
fare_array = df['Fare'].to_numpy()

for cls in [1,2,3]:
    filter_cls = pclass_array == cls  #filter_cls = 4 == 1,2,3 : false, 
    class_fares = fare_array[filter_cls]
    print(f'Class{cls} - Avg Fare:{np.mean(class_fares):.2f} | Max: {np.max(class_fares):.2f} | Min: {np.min(class_fares):.2f}')
```

```python
class_counts = df['Pclass'].value_counts().sort_index()

print(class_counts)


class_lables = ['1st class','2nd class', '3rd class']

plt.figure(figsize=(8,4))

plt.bar(class_lables, class_counts, color =['gold','silver','blue'], width=0.5)

plt.title('Number of passengers by class')
plt.xlabel('Passengers class')
plt.ylabel('Number od Passengers')

for i, count in enumerate(class_counts):
    # plt.text(i,count + 2, str(count), ha = 'center', fontsize = 12, fontweight = 'bold')
    plt.text(i, count + 2, f'{count}', ha='center', fontweight = 'bold')
```

```python
# Bar chart — Survival rate by class (in percentage)

survival_by_class  = df.groupby('Pclass')['Survived'].mean() * 100
class_labels = ['1st Class', '2nd Class', '3rd Class']
colors = ['#2ecc71', '#3498db', '#e74c3c']

plt.figure(figsize=(8,5))
plt.bar(class_labels,survival_by_class, color = colors, width = 0.5)
plt.title('Survival Rate by Passengers Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate(%)')
plt.ylim(0,100)

for i, val in enumerate(survival_by_class.values):
    plt.text(i, val + 1, f'{val:.1f}%', ha='center', fontsize=12, fontweight = 'bold')

plt.show()

```

```python
#horizontal bar chart - survival rate by boarding port
# Cherbourg (C), Queenstown (Q), Southampton (S)

port_survival = df.groupby('Embarked')['Survived'].mean() * 100

port_labels = ['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)']
plt.figure(figsize=(8,4))
plt.barh(port_labels, port_survival, color = 'coral', height = 0.5)
plt.title('Survival Rate By Boarding Port')
plt.xlabel('Survival Rate(%)')
plt.ylabel('Boarding Port')
plt.xlim(0,100)

for i, val in enumerate(port_survival.values):
    plt.text(val +1, i, f'{val:.1f}%', va = 'center', fontsize = 12)

plt.show()
```
