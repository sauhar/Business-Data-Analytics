```python
hours = 13
base_fare = 150

if hours >= 8 and hours < 11:
    increase = 1.5
    time_type = "Morning Peak"
elif hours >=16 and hours < 19:
    increase = 1.5
    time_type = "Evening Peak"
else:
    increase = 1.0
    time_type = "Off Peak"

final_fare = base_fare * increase
print(f"Time Type: {time_type}, Final Fare: {final_fare}")
```

```python
# use function to calculate items price with vat(13%)

def calcaulte_vat(price):
    vate_rate = 0.13
    vat_amount = price * vate_rate
    total_price = price + vat_amount
    return total_price

item_1 = calcaulte_vat(500) 
item_2 = calcaulte_vat(250) 
item_3 = calcaulte_vat(200) 

print(item_1)
print(item_2)
 
```
