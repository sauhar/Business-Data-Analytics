## 2. 📦 Variables — Storing Information

A **variable** is like a box with a label.
You put information inside the box, and use the label to find it later.

## 3. 🔢 Data Types — What Kind of Data?

Python stores different **types** of data:

| Type | Example | Description |
|------|---------|-------------|
| `int` | `100` | Whole numbers |
| `float` | `99.50` | Numbers with decimal |
| `str` | `"Daraz"` | Text (called a string) |
| `bool` | `True` / `False` | Yes or No |

Use `type()` to check what type a variable is.

````python
# Let's check the data type of our variables

product_name = "Wai Wai Noodles"    # text
price = 25.50                         # decimal number
quantity = 200                        # whole number
in_stock = True                       # yes/no

print(type(product_name))   # str = text
print(type(price))          # float = decimal
print(type(quantity))       # int = whole number
print(type(in_stock))       # bool = True/False
````

## 4. ➕ Doing Math with Python

Python can work as a powerful calculator for your business!

| Symbol | Meaning | Example |
|--------|---------|----------|
| `+` | Addition | `500 + 300` |
| `-` | Subtraction | `500 - 200` |
| `*` | Multiplication | `50 * 10` |
| `/` | Division | `100 / 4` |
| `**` | Power (exponent) | `2 ** 3` = 8 |
| `%` | Remainder | `10 % 3` = 1 |

````python
# eSewa transaction calculation example

transaction_amount = 5000       # Customer sent Rs. 5000
service_charge_rate = 0.01      # 1% service charge
service_charge = transaction_amount * service_charge_rate
amount_received = transaction_amount - service_charge

print(f"Transaction Amount: Rs. {transaction_amount}")
print(f"Service Charge (1%): Rs. {service_charge}")
print(f"Amount Received: Rs. {amount_received}")
````

````python
# Profit and Loss Calculation — Nepal Telecom example

revenue = 450000000    # Rs. 45 crore revenue
expenses = 320000000   # Rs. 32 crore expenses

profit = revenue - expenses
profit_margin = (profit / revenue) * 100    # profit as a percentage

print(f"Revenue:       Rs. {revenue:,}")
print(f"Expenses:      Rs. {expenses:,}")
print(f"Profit:        Rs. {profit:,}")
print(f"Profit Margin: {profit_margin:.2f}%")

# Note: :, adds comma separators to big numbers (e.g., 1,000,000)
# Note: :.1f rounds to 1 decimal place
````

## 5. 📝 Working with Text (Strings)

In Python, text is called a **string**. It is always written inside quotes.

You can do many useful things with text!

````python
# String basics

customer_name = "Sita Sharma"
email = "sita.sharma@gmail.com"
address = "   Lalitpur, Nepal   "   # notice extra spaces

# Change to UPPERCASE
print(customer_name.upper())

# Change to lowercase
print(customer_name.lower())

# Remove extra spaces
print(address.strip())

# Count characters
print(len(customer_name))

# Check if something is in the text
print("Sharma" in customer_name)    # True
print("Thapa" in customer_name)     # False
````

````python
# Joining strings (concatenation)

first_name = "Ram"
last_name = "Thapa"

# Method 1: Using +
full_name = first_name + " " + last_name
print(full_name)

# Method 2: Using f-string (recommended!)
full_name = f"{first_name} {last_name}"
print(f"Customer Full Name: {full_name}")
````

## 6. 📋 Lists — Storing Multiple Items

A **list** stores multiple items in one variable — like a shopping list or product catalog.

```python
my_list = [item1, item2, item3]
```

⚠️ **Important:** In Python, counting starts from **0**, not 1.

````python
# List of products in a Daraz store

products = ["Mobile Phone", "Laptop", "Headphones", "Charger", "Smart Watch"]

# Print all products
print(products)

# Get the first product (index 0)
print(products[0])

# Get the last product
print(products[-1])

# How many products do we have?
print(f"Total Products: {len(products)}")


mixed = [10, "Hello", 3.14, True]  # int, str, float, bool
print(mixed)
````

````python
# Adding and removing items from a list

products = ["Mobile Phone", "Laptop", "Headphones"]
print(f"Before: {products}")

# Add a new product
products.append("Tablet")
print(f"After adding Tablet: {products}")

# Remove a product
products.remove("Headphones")
print(f"After removing Headphones: {products}")

# Sort alphabetically
products.sort()
print(f"Sorted: {products}")
````

## 7. 📖 Dictionaries — Key & Value Pairs

A **dictionary** stores data in **key: value** pairs — like a real dictionary where each word (key) has a meaning (value).

Perfect for storing structured business data!

## 8. 🚦 If / Else — Making Decisions

Python can make decisions based on conditions — just like business rules!

```python
if condition:
    # do this if condition is True
else:
    # do this if condition is False
```

**Comparison operators:**

| Symbol | Meaning |
|--------|---------|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal |
| `<=` | Less than or equal |

````python
# Loan approval system — like a bank in Nepal
#Check if
#✅ If true → run it and STOP
#If false → check elif
#✅ If true → run it and STOP
#If both false → run else

monthly_income = 45000   # in Rs.
credit_score = 720

if monthly_income >= 40000 and credit_score >= 700:
    print("✅ Loan Approved! You qualify for a home loan.")
elif monthly_income >= 25000 and credit_score >= 600:
    print("⚠️  Partial Approval — You qualify for a smaller loan.")
else:
    print("❌ Loan Rejected — Income or credit score too low.")
````

````python
# Pathao ride pricing — peak hour vs normal

hour = 8          # 8 AM
base_fare = 100   # Rs. 100 base

if hour >= 7 and hour <= 10:       # Morning peak hours
    increase = 1.5
    time_type = "Morning Peak"
elif hour >= 17 and hour <= 20:    # Evening peak hours
    increase = 1.8
    time_type = "Evening Peak"
else:
    increase = 1.0
    time_type = "Normal"

final_fare = base_fare * surge
print(f"Time Type: {time_type}")
print(f"Surge Multiplier: {surge}x")
print(f"Final Fare: Rs. {final_fare}")
````

## 9. 🔁 Loops — Repeating Tasks Automatically

Loops let Python do repetitive tasks for you — automatically!

Instead of printing 100 items one by one, you can write a loop.

Two types:
- `for` loop — repeat a fixed number of times
- `while` loop — repeat until a condition is false

````python
# FOR LOOP — Go through a list of items

# Print all product names from a list
products = ["Wai Wai", "Goldstar Shoes", "CG Electronics", "Dabur Nepal", "Ncell SIM"]

print("--- Product List ---")
for product in products:
    print(f"  • {product}")
````

````python
try_clear = ["Apple", "Banana", "Mango"]
print(try_clear)

# try_clear.clear()
# print(try_clear)
````

````python
# Loop with index number using enumerate()

products = ["Wai Wai", "Goldstar Shoes", "CG Electronics"]

print("--- Numbered Product List ---")
for a, product in enumerate(products, start=1):
    print(f"{a}. {product}")
````

````python
#print(len(daily_sales))

word = "hello world"
print(len(word))  # Output: 5
````

````python
# Calculating total sales using a loop

daily_sales = [12000, 15500, 9800, 22000, 18300, 25000, 11000]  # 7 days of sales

total = 0
for sale in daily_sales:
    total = total + sale   # Add each day's sale to total

average = total / len(daily_sales) #number of items in the list (7 in this case)

print(f"Total Weekly Sales: Rs. {total:,}")
print(f"Average Daily Sales: Rs. {average:,.0f}")
````

````python
# WHILE LOOP — Repeat until something happens

# Counting down a delivery timer
days_left = 5

print("Daraz Delivery Countdown:")
while days_left > 0:
    print(f"  {days_left} day(s) remaining...")
    days_left = days_left - 1

print("📦 Your package has arrived!")
````

## 10. 🔧 Functions — Reusable Code Blocks

A **function** is a block of code you write once and use many times.

Think of it like a **formula in Excel** — you define it once, then call it whenever needed.

```python
def function_name(input):
    # do something
    return result
```

````python
# Function to calculate VAT (13% in Nepal)

def calculate_vat(price):
    vat_rate = 0.13     # 13%
    vat_amount = price * vat_rate
    total_price = price + vat_amount
    return total_price

# Now we can use this function for any price!
item1 = calculate_vat(500)
item2 = calculate_vat(2000)
item3 = calculate_vat(15000)

print(f"Rs. 500 + VAT = Rs. {item1}")
print(f"Rs. 2000 + VAT = Rs. {item2}")
print(f"Rs. 15000 + VAT = Rs. {item3}")
````

````python
# Function with multiple inputs

def calculate_discount(original_price, discount_percent):
    discount_amount = original_price * (discount_percent / 100)
    final_price = original_price - discount_amount
    return final_price

# Dashain sale at Bhatbhateni!
shirt_price = calculate_discount(1500, 20)   # 20% off
tv_price = calculate_discount(55000, 15)     # 15% off

print(f"Shirt after 20% discount: Rs. {shirt_price}")
print(f"TV after 15% discount: Rs. {tv_price}")
````

````python
# Function to generate a customer receipt

def print_receipt(customer_name, product, quantity, unit_price):
    subtotal = quantity * unit_price
    vat = subtotal * 0.13
    total = subtotal + vat

    print("=" * 35)
    print("      BHATBHATENI SUPERSTORE")
    print("         Kathmandu, Nepal")
    print("=" * 35)
    print(f"Customer: {customer_name}")
    print(f"Product:  {product}")
    print(f"Qty:      {quantity} x Rs.{unit_price}")
    print("-" * 35)
    print(f"Subtotal: Rs. {subtotal:,.2f}")
    print(f"VAT 13%:  Rs. {vat:,.2f}")
    print(f"TOTAL:    Rs. {total:,.2f}")
    print("=" * 35)

# Generate a receipt!
print_receipt("Ram Thapa", "Basmati Rice (5kg)", 3, 450)
````

## 11. 💼 Mini Business Exercise

Now let's put it all together!

**Scenario:** You manage a small electronics shop in Kathmandu.
You have a list of products with their prices and stock.
Your task: Find which products are low on stock and print a report.

> **Low stock = quantity less than 5 items**

````python
# Inventory Management — Mini Exercise

# Step 1: Create inventory as a list of dictionaries
inventory = [
    {"product": "Samsung Mobile",   "price": 45000, "quantity": 8},
    {"product": "HP Laptop",        "price": 85000, "quantity": 3},
    {"product": "JBL Headphone",    "price": 8000,  "quantity": 15},
    {"product": "Xiaomi Power Bank","price": 3500,  "quantity": 2},
    {"product": "USB-C Charger",    "price": 1200,  "quantity": 4},
    {"product": "Smart Watch",      "price": 12000, "quantity": 10},
]

# Step 2: Find low stock items
print("📦 INVENTORY REPORT — Electronics Shop Kathmandu")
print("=" * 50)
print(f"{'Product':<25} {'Qty':>5} {'Status':<15}")
print("-" * 50)

low_stock_items = []   # Empty list to collect alerts

for item in inventory:
    if item["quantity"] < 5:
        status = "⚠️  LOW STOCK"
        low_stock_items.append(item["product"])
    else:
        status = "✅ OK"

    print(f"{item['product']:<25} {item['quantity']:>5}  {status}")

print("-" * 50)
print(f"\n🔔 Reorder needed for: {', '.join(low_stock_items)}")
````

````python
# Bonus: Calculate total inventory value

total_value = 0
for item in inventory:
    value = item["price"] * item["quantity"]
    total_value = total_value + value

print(f"💰 Total Inventory Value: Rs. {total_value:,}")
````
