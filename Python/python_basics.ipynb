## 4. ➕ Doing Math with Python

Python can work as a powerful calculator for your business!

| Symbol | Meaning | Example |
|--------|---------|----------|
| `+` | Addition | `500 + 300` |
| `-` | Subtraction | `500 - 200` |
| `*` | Multiplication | `50 * 10` |
| `/` | Division | `100 / 4` |
| `**` | Power (exponent) | `2 ** 3` = 8 |
| `%` | Remainder | `10 % 3` = 1 |

````python
# eSewa transaction calculation example

transaction_amount = 5000       # Customer sent Rs. 5000
service_charge_rate = 0.01      # 1% service charge
service_charge = transaction_amount * service_charge_rate
amount_received = transaction_amount - service_charge

print(f"Transaction Amount: Rs. {transaction_amount}")
print(f"Service Charge (1%): Rs. {service_charge}")
print(f"Amount Received: Rs. {amount_received}")
````

````python
# Profit and Loss Calculation — Nepal Telecom example

revenue = 450000000    # Rs. 45 crore revenue
expenses = 320000000   # Rs. 32 crore expenses

profit = revenue - expenses
profit_margin = (profit / revenue) * 100    # profit as a percentage

print(f"Revenue:       Rs. {revenue:,}")
print(f"Expenses:      Rs. {expenses:,}")
print(f"Profit:        Rs. {profit:,}")
print(f"Profit Margin: {profit_margin:.2f}%")

# Note: :, adds comma separators to big numbers (e.g., 1,000,000)
# Note: :.1f rounds to 1 decimal place
````

## 5. 📝 Working with Text (Strings)

In Python, text is called a **string**. It is always written inside quotes.

You can do many useful things with text!

````python
# String basics

customer_name = "Sita Sharma"
email = "sita.sharma@gmail.com"
address = "   Lalitpur, Nepal   "   # notice extra spaces

# Change to UPPERCASE
print(customer_name.upper())

# Change to lowercase
print(customer_name.lower())

# Remove extra spaces
print(address.strip())

# Count characters
print(len(customer_name))

# Check if something is in the text
print("Sharma" in customer_name)    # True
print("Thapa" in customer_name)     # False
````

````python
print("Hello\nWorld")  # \n creates a new line
print("Price:\tRs. 100")  # \t adds a tab space
````

````python
# Joining strings (concatenation)

first_name = "Ram"
last_name = "Thapa"

# Method 1: Using +
full_name = first_name + " " + last_name
print(full_name)

# Method 2: Using f-string (recommended!)
full_name = f"{first_name} {last_name}"
print(f"Customer Full Name: {full_name}")
````

````python
# Splitting and joining strings — very useful for data cleaning!

# Split a sentence into a list of words
address = "Thamel, Kathmandu, Bagmati, Nepal"
parts = address.split(",")  # split wherever there is a comma
print(parts)
print(f"City: {parts[1].strip()}") #.strip() removes leading and trailing whitespace from a string.

# Join a list back into a string
tags = ["sales", "marketing", "nepal", "b2c"]
joined = " | ".join(tags)
print(joined)
````

## 🔵 Sets — Unique Items Only

A **set** automatically removes duplicates. It only keeps **unique values**.
Very useful in business for finding distinct customers, unique cities, etc.

```python
my_set = {item1, item2, item3}
```

````python
# Find unique cities from customer orders

customer_cities = ["Kathmandu", "Pokhara", "Kathmandu", "Biratnagar",
                   "Pokhara", "Lalitpur", "Kathmandu", "Butwal", "Lalitpur"]

# Convert to set — duplicates disappear!
unique_cities = set(customer_cities)
print(f"\nUnique cities: {unique_cities}")
print(f"Number of unique cities: {len(unique_cities)}")
````

## A3. Slicing — Getting a Portion of a List

You can get a **range of items** using slicing:
```python
list[start : end]     # end is NOT included!
list[start : end : step]
```

````python
products = ["Laptop", "Mobile", "Tablet", "Headphone", "Charger", "Smart Watch", "Camera"]

# Get first 3 items
print(products[0:3])     # index 0, 1, 2 (3 is NOT included)

# Get items from index 2 to 5
print(products[2:5])

# Get from beginning up to index 4
print(products[:4])

# Get from index 3 to the end
print(products[3:])

# Get every second item (step = 2)
print(products[::2])  #skip one value and get the next one

# Reverse the list
print(products[::-1])
````

## 8. 🚦 If / Else — Making Decisions

Python can make decisions based on conditions — just like business rules!

```python
if condition:
    # do this if condition is True
else:
    # do this if condition is False
```

**Comparison operators:**

| Symbol | Meaning |
|--------|---------|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal |
| `<=` | Less than or equal |

````python
# Loan approval system — like a bank in Nepal
#Check if
#✅ If true → run it and STOP
#If false → check elif
#✅ If true → run it and STOP
#If both false → run else

num = 0

# Check the number
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
````

````python
# Pathao ride pricing — peak hour vs normal

hour = 8          # 8 AM
base_fare = 100   # Rs. 100 base

if hour >= 7 and hour <= 10:       # Morning peak hours
    increase = 1.5
    time_type = "Morning Peak"
elif hour >= 17 and hour <= 20:    # Evening peak hours
    increase = 1.8
    time_type = "Evening Peak"
else:
    increase = 1.0
    time_type = "Normal"

final_fare = base_fare * increase
print(f"Time Type: {time_type}")
print(f"Increase Multiplier: {increase}x")
print(f"Final Fare: Rs. {final_fare}")
````

## 9. 🔁 Loops — Repeating Tasks Automatically

Loops let Python do repetitive tasks for you — automatically!

Instead of printing 100 items one by one, you can write a loop.

Two types:
- `for` loop — repeat a fixed number of times
- `while` loop — repeat until a condition is false

````python
# FOR LOOP — Go through a list of items

# Print all product names from a list
products = ["Wai Wai", "Goldstar Shoes", "CG Electronics", "Dabur Nepal", "Ncell SIM"]

print("--- Product List ---")
for product in products:
    print(product)
````

````python
# Loop with index number using enumerate()

products = ["Wai Wai", "Goldstar Shoes", "CG Electronics"]

for index, product in enumerate(products, start = 1):
    print(index, product)
````

````python
#print(len(daily_sales))

word = "hello world"
print(len(word))  # Output: 5
````

````python
# Calculating total sales using a loop

daily_sales = [12000, 15500, 9800, 22000, 18300, 25000, 11000]  # 7 days of sales

total = 0
for sale in daily_sales:
    total = total + sale   # Add each day's sale to total

average = total / len(daily_sales) #number of items in the list (7 in this case)

print(f"Total Weekly Sales: Rs. {total:,}")
print(f"Average Daily Sales: Rs. {average:,.0f}")
````

````python
# WHILE LOOP — Repeat until something happens

# Counting down a delivery timer
days_left = 5

print("Daraz Delivery Countdown:")
while days_left > 0:
    print(days_left)
    days_left = days_left - 1

print("📦 Your package has arrived!")
````

## 10. 🔧 Functions — Reusable Code Blocks

A **function** is a block of code you write once and use many times.

```python
def function_name(input):
    # do something
    return result
```

````python
# del function_name =  to del function

def add_numbers(a, b):   #a and b are called parameters
    return a + b
result = add_numbers(5, 3)
print(result)
````

````python
def multiply(a, b):
    return a * b

def square(n):
    return n ** 2

x = 4
y = 5

print("Multiplication:", multiply(x, y))
print("Square of x:", square(x))
print("Square of y:", square(y))
````

````python
numbers = [5,4,6,7,23]

numbers.sort(reverse = True)

print(numbers)
````

````python
# Function to calculate VAT (13% in Nepal)

def calculate_vat(price):
    vat_rate = 0.13     # 13%
    vat_amount = price * vat_rate
    total_price = price + vat_amount
    return total_price

# Now we can use this function for any price!
item1 = calculate_vat(500)
item2 = calculate_vat(2000)
item3 = calculate_vat(15000)

print(f"Rs. 500 + VAT = Rs. {item1}")
print(f"Rs. 2000 + VAT = Rs. {item2}")
print(f"Rs. 15000 + VAT = Rs. {item3}")
````

````python
# Function with multiple inputs

# Function that takes three numbers and returns their sum
def sum_three_numbers(a, b, c):
    return a + b + c

# Get input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

# Call the function with the inputs
result = sum_three_numbers(num1, num2, num3)

# Display the result
print("The sum of the three numbers is:", result)
````

## 11. 🛡️ Try / Except — Handling Errors Gracefully

Errors happen! Python lets you **catch errors** and handle them nicely instead of crashing.

Think of it like a business backup plan — if Plan A fails, go to Plan B.

```python
try:
    # code that might cause an error
except SomeError:
    # what to do if error happens
```

| Error Type          | When it happens                                                               |
| ------------------- | ----------------------------------------------------------------------------- |
| `ZeroDivisionError` | When you divide by zero                                                       |
| `ValueError`        | When a function gets a value of the wrong type (like converting "abc" to int) |
| `TypeError`         | When you use the wrong type in an operation (e.g., `"5" + 2`)                 |
| `FileNotFoundError` | When trying to open a file that doesn’t exist                                 |
| `IndexError`        | When accessing an invalid index in a list                                     |
| `KeyError`          | When accessing a dictionary with a key that doesn’t exist                     |
| `Exception`         | The base class for all errors (catches everything)                            |


````python
try:
    result = 10 / 0
    print(result)
except ZeroDivisionError:
    print("You cannot divide by zero!")
````

````python
numbers = [5, 2, 9, 1]

numbers.sort(reverse=True)

print(numbers)
````

````python
try:
    num = int("abc")
    print(num)
except ValueError:
    print("Invalid input! Please enter a number.")
````

````python
try:
    result = "5" + 2
    print(result)
except TypeError:
    print("You cannot add string and integer!")
````

````python
try:
    numbers = [1, 2, 3]
    print(numbers[3])
except IndexError:
    print("Index out of range!")
````

````python
try:
    student = {"name": "Ram", "age": 20}
    print(student["address"])
except KeyError:
    print("Key not found in dictionary!")
````

````python
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
    print( result)
except ZeroDivisionError:
    print("Oops! You cannot divide by zero.")
````
