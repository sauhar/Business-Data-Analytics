# Cleaning

```python
# df['PassengerId'].duplicated().any()
df['Survived'].duplicated().any()
```

```python
# Add a new column — Age Group (Child, Adult, Senior).

def age_group(age):
    if age < 18:  return 'Child'
    elif age < 60: return 'Adult'
    else:          return 'Senior'

df['AgeGroup'] = df['Age'].apply(age_group) #Takes each value in df['Age'], Passes it into age_group(), Stores result in a new column AgeGroup
df['AgeGroup'].value_counts() #gives records count on the basic of category
```

### Cleaning 1: Fix Missing Age Values

```python
# First, see exactly how many ages are missing
missing_age = df['Age'].isnull().sum()
print('Number of missing ages:', missing_age)
```

```python

# Fill missing Age values with the average age.

median_age = df['Age'].median()           # calculate average
print(median_age)                       # ~30.27

df['Age'] = df['Age'].fillna(median_age)  # fill NaN with average
df['Age'].isnull().sum()             # → 0 (no more missing!)
```

```python
# Round Age to nearest whole number
# Decimal ages like 34.5 or 22.5 are estimated — we round them for cleanliness
# Note: Infant ages like 0.17 (2-month-old baby) also get rounded to 0

df['Age'] = df['Age'].round(0)

print('Age column after rounding:')
print(df['Age'].head(10))
```

### Cleaning 2: Fix Missing Fare Value

```python
# Check missing fare
print('Missing fare values:', df['Fare'].isnull().sum())
```

```python
# Fill the missing fare with the median fare
# median() calculates the middle value

median_fare = df['Fare'].median()
print('Median fare:', median_fare)

df['Fare'] = df['Fare'].fillna(median_fare)

print('Missing fare after fix:', df['Fare'].isnull().sum())
```

```python
# Round Fare to 2 decimal places — cleaner to read
df['Fare'] = df['Fare'].round(2)

print('Sample fares after rounding:')
print(df['Fare'].head(10))
```

### Cleaning 3: Fix the Cabin Column (78% missing)

```python
# 327 out of 418 cabin values are missing (78%)
# We cannot guess 78% of missing values — that would be making up data
# Instead: create a simple Yes/No column — did they have a cabin or not?

# notna() = True if there IS a value, False if empty
df['Has_Cabin'] = df['Cabin'].notna()

print('Has Cabin — True means cabin is known:')
print(df['Has_Cabin'].value_counts())
```

```python
# Also extract the deck letter from Cabin
# Cabin codes like 'B45' start with a letter = the deck of the ship
# A = top deck, G = lower deck
# If cabin is missing, we write 'Unknown'

def get_deck(cabin):
    if pd.isnull(cabin):       # if cabin is empty
        return 'Unknown'
    return cabin[0]            # first letter = deck name

df['Deck'] = df['Cabin'].apply(get_deck)

print('Deck distribution:')
print(df['Deck'].value_counts())
```

```python
# 22 names have extra " characters — like a typing mistake
# Example: Dean, Miss. Elizabeth Gladys Millvina"" (extra quotes at end)

# str allows us to perform string operations on each element of a Series individually.
# str.replace() finds a character and removes it
# str.strip() removes extra spaces from start and end

df['Name_raw'] = df['Name'] 

df['Name'] = df['Name'].str.replace('"', '', regex=False)
df['Name'] = df['Name'].str.strip()


print(df['Name'].head(5).tolist())
```

```python
df.rename(columns={
    'Name_raw': 'RawName',
    'Has_Cabin': 'HasCabin',
}, inplace=True)  # inplace=True Apply the change directly to the original DataFrame instead of creating a new one. In default its False

#to drop the unwanted column:
    # df = df.drop('Fare',axis = 1)
    # df = df.drop(['Fare', 'Cabin'], axis=1)
```

```python
# Create FamilySize column
# SibSp = number of brothers/sisters/husband/wife on board
# Parch = number of parents/children on board
# +1 because we include the passenger themselves

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

print('Family size of first 10 passengers:')
print(df[['Name', 'SibSp', 'Parch', 'FamilySize']].head(10))
```

```python
# Create a simple label for family size
# Solo = alone | Small = 2 to 4 | Large = 5 or more

def family_label(size):
    if size == 1:
        return 'Solo'
    elif size <= 4:
        return 'Small'
    else:
        return 'Large'

df['FamilyType'] = df['FamilySize'].apply(family_label)

print('Family type count:')
print(df['FamilyType'].value_counts())
```

```python
df.head()
```

```python
# Create Fare Category column
# pd.cut() takes a continuous numeric column (Fare) and splits it into intervals (like buckets or brackets).
# Think of it like: Economy / Business / First Class pricing

# | Fare Range | Category |
# | ---------- | -------- |
# | 0–10       | Budget   |
# | 10–30      | Standard |
# | 30–100     | Premium  |
# | 100–600    | Luxury   |


df['FareCategory'] = pd.cut(
    df['Fare'],
    bins   = [0, 10, 30, 100, 600], # right-inclusive → (a, b] like "(0, 10] → includes 10, excludes 0"
    labels = ['Budget', 'Standard', 'Premium', 'Luxury'],
    include_lowest = True  # This ensures the lowest value (0) is included in the first bin.
)

print('Fare category count:')
print(df['FareCategory'].value_counts())
```

```python
# Create a column: Was the passenger traveling alone?
# If FamilySize is 1, they were alone

def check_alone(x):
    if x == 1:
        return 'Yes'
    else:
        return 'No'

df['Alone'] = df['FamilySize'].apply(check_alone)
```

```python
# Pivot table: rows = Gender, columns = Class
# Values = survival rate (how many % survived)
# Think of it like a report card — rows on one side, columns on other

pivot = pd.pivot_table(
    df,
    index   = 'Sex',
    columns = 'Pclass',
    values  = 'Survived',
    aggfunc = 'mean'
) * 100

pivot = pivot.round(1)
pivot.columns = ['Class 1 %', 'Class 2 %', 'Class 3 %']

print('Survival rate % by Gender and Class:')
pivot
```
